```python
import cv2
import numpy as np
from keras.models import model_from_json
```

```python
e_dict={0:"Angry",1:"Disgust",2:"fearful",3:"happy",4:"neutral",5:"sad",6:"suprised" }
```

```python
json_file=open(r"C:\Users\snigdh chamoli\OneDrive\Desktop\emotionde\model.json",'r')
load_model=json_file.read()
json_file.close()
e_model=model_from_json(load_model)
```

```python
e_model.load_weights(r"C:\Users\snigdh chamoli\OneDrive\Desktop\emotionde\model.h5")
```

```python
face_detector=cv2.CascadeClassifier(r'C:\Users\snigdh chamoli\OneDrive\Desktop\emotionde\haarcascade_frontalcatface.xml')
# cap=cv2.VideoCapture(0)
cap=cv2.VideoCapture(r"C:\Users\snigdh chamoli\OneDrive\Desktop\emotionde\production_id_3700029 (2160p).mp4")

while True:
    ret,frame=cap.read()
    frame=cv2.resize(frame,(400,500))
    if not ret:
        break
    
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_detector.detectMultiScale(gray,scaleFactor=1.05,minNeighbors=5)
    for(x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),4)
        roi_gray=gray[y:y+h,x:x+w]
        cropped_img=np.expand_dims(np.expand_dims(cv2.resize(roi_gray,(48,48)),-1),0)
        
        
        prediction=e_model.predict(cropped_img)
        max_Index=int (np.argmax(prediction))
        cv2.putText(frame,e_dict[max_Index],(x+5,y-20),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2,cv2.LINE_AA)
    cv2.imshow('SENTIMENT ANALYSIS THROUGH FACE DETECTION',frame)
    if cv2.waitKey(1) & 0xFF ==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
